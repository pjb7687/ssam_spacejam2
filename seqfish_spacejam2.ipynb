```python
from collections import defaultdict
import numpy as np

minmax_x = [9999999999999, -9999999999999]
minmax_y = [9999999999999, -9999999999999]
minmax_z = [9999999999999, -9999999999999]

pos_dic = defaultdict(lambda: [])
pos_dic_3d = defaultdict(lambda: [])
with open("data/seqFISH_spots_with_anatomy_microns.csv") as f:
    print(f.readline())
    for line in f:
        entries = line.rstrip().split(',')
        x = float(entries[7])
        y = float(entries[9])
        z = float(entries[11])
        g = entries[3]
        pos_dic[g].append([x, y])
        pos_dic_3d[g].append([x, y, z])
        if x < minmax_x[0]:
            minmax_x[0] = x
        if x > minmax_x[1]:
            minmax_x[1] = x
        if y < minmax_y[0]:
            minmax_y[0] = y
        if y > minmax_y[1]:
            minmax_y[1] = y
        if z < minmax_z[0]:
            minmax_z[0] = z
        if z > minmax_z[1]:
            minmax_z[1] = z
for g in pos_dic:
    pos_dic[g] = np.array(pos_dic[g])
for g in pos_dic_3d:
    pos_dic_3d[g] = np.array(pos_dic_3d[g])
```

```python
all_genes = sorted(pos_dic.keys())
mrna_loci = [pos_dic[gene] for gene in all_genes]
```

```python
import pickle
with open("scrna_sigs.pkl", "rb") as f:
    scrna_clusters, scrna_genes, scrna_full_centroids, scrna_colors = pickle.load(f)
```

```python
scrna_full_centroids = np.pad(scrna_full_centroids, [(0, 0), (0, 6)], mode='constant')
```

```python
not_found_genes = [
    "9330158h04rik",
    "A330009n23rik",
    "A830009l08rik",
    "D030045p18rik",
    "D1ertd622e",
    "Loc432842",
]
scrna_genes = list(scrna_genes)
for g in not_found_genes:
    scrna_genes.append(g)
scrna_genes = np.array(scrna_genes)
```

```python
gene_conv_dic = {
    "Ai593442": "AI593442",
    "Aw551984": "AW551984",
    "Bc006965": "BC006965",
    "GM13848": "Gm13848",
}
```

```python
for f, t in gene_conv_dic.items():
    ds.genes = np.where(ds.genes == f, t, ds.genes)
for gidx in range(len(ds.genes)):
    ds.genes[gidx] = ds.genes[gidx][0].upper() + ds.genes[gidx][1:]
```

```python
selected_genes_idx = [list(scrna_genes).index(g) for g in ds.genes]
scrna_uniq_clusters = np.unique(scrna_clusters)
scrna_centroids = scrna_full_centroids[:, selected_genes_idx]
```

```python
sf_scrna_corrs = np.zeros((len(scrna_centroids), len(ds.centroids)))
for i, scrna_centroid in enumerate(scrna_centroids):
    for j, centroid in enumerate(ds.centroids):
        sf_scrna_corrs[i, j] = ssam.utils.corr(scrna_centroid, centroid)
sf_scrna_max_corr = np.max(sf_scrna_corrs, axis=0)
sf_scrna_max_corr_idx = np.argmax(sf_scrna_corrs, axis=0)
sf_scrna_cluster_labels = np.array(scrna_clusters)[sf_scrna_max_corr_idx]
```

```python
sf_scrna_colors = [scrna_colors[scrna_clusters.index(cl)] for cl in sf_scrna_cluster_labels]
```

```python
plt.figure(figsize=[15,15])
_ = ds.plot_celltypes_map(colors=sf_scrna_colors)
```

```python
analysis.bin_celltypemaps(step=10, radius=100)
analysis.find_domains(n_clusters=20, merge_remote=False, merge_thres=0.8, norm_thres=1500)
plt.figure(figsize=[15, 15])
ds.plot_domains(rotate=0)
```
